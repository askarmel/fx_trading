```python
import pandas as pd
import plotly.graph_objects as go
import datetime as dt
```

```python
pair = "EUR_USD"
granularity = "H4"
df = pd.read_pickle(f"../data/{pair}_{granularity}.pkl")
```

```python
df_plot = df.iloc[:100]
```

```python
fig = go.Figure()
fig.add_trace(go.Candlestick(
    x = df_plot.time,
    open = df_plot.mid_o,
    close = df_plot.mid_c,
    low = df_plot.mid_l,
    high = df_plot.mid_h,
))

fig.show()
```
